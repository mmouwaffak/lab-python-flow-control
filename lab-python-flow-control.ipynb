## Exercise: Managing Customer Orders Optimized

```python

#1. Look at your code from the lab data structures, and improve repeated code with loops:

products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = {}

for product in products:
    quantity = int(input(f"Enter the available quantity for {product}: "))  #f strings: to format selected parts of a string
    inventory[product] = quantity
    print(f"{product}: {inventory[product]}")


customer_orders = set()
print("Current customer orders (empty set):", customer_orders)

#2. Instead of asking the user to input the name of three products that a customer wants to order, do the following:
# a. Prompt the user to enter the name of a product that a customer wants to order:
# b. Add the product name to the "customer_orders" set:
# c. Ask the user if they want to add another product (yes/no):
# d. Continue the loop until the user does not want to add another product:

while True:
    product_name = input("Enter the name of a product to add to the order: ").strip().lower()

    if product_name in products:
        customer_orders.add(product_name)
        print(f"{product_name} added to the order.")
    else:
        print("Invalid product, this item will be ignored.")

    another_product = input("Do you want to add another product? (yes/no): ").strip().lower()
    if another_product not in ("yes", "y", "Y", "YES", "Yes"):
        break

 
print("\nProducts in the customer's order (set):") # \n to create a new line
for item in customer_orders:
    print("-", item)

# Order statistics:
total_products_ordered = len(customer_orders)

# % of products ordered compared to the total number of different products:
percentage_ordered = (total_products_ordered / len(products)) * 100

order_status = (total_products_ordered, percentage_ordered)

# Print the order statistics:
print("\n*Order Statistics:*")
print(f"Total Products Ordered (distinct): {order_status[0]}")
print(f"Percentage of Products Ordered: {order_status[1]:.2f}%")

# 3. Instead of updating the inventory by subtracting 1 from the quantity of each product, 
# only do it for  the products that were ordered (those in "customer_orders"):

for ordered_product in customer_orders:
    if inventory[ordered_product] > 0:
        inventory[ordered_product] -= 1
    else:
        print(f"Warning: no inventory left for '{ordered_product}'.")


print("\nUpdated Inventory")
for product in products:
    print(f"{product}: {inventory[product]}")
```
