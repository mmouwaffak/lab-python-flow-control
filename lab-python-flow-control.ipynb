```python
# answers:

print("\nInventory")

# 1. Define a list called product:
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# 2. Create an empty inventory dictionary:
inventory = {}

# 3. Ask the user to enter the quantity available for each product

for product in products:
    quantity = int(input(f"Enter the available quantity for '{product}': "))
    inventory[product] = quantity
    print(f"{product}: {inventory[product]}")
# 4. Create an empty set for customer orders
customer_orders = set()
print(customer_orders)

# 5. Ask the user to input three products the customer wants to order:

for i in range(1, 4):
    product_name = input(f"Enter the name of product #{i}: ").strip().lower()     
    #F-strings are string literals prefixed with 'f' or 'F' that contain expressions inside curly braces {}.
    # The user has to type a product from the list (t-shirt, mug, hat, book, keychain)
    if product_name in products:
        customer_orders.add(product_name)
    else:
        print("Invalid product, this item will be ignored.")

# 6. Print the products in the customer_orders set:
print("Products in the customer's order (set):")       
for item in customer_orders:
    print("- ", item)

# 7. Calculate order statistics:
total_products_ordered = len(customer_orders)

# Percentage of products ordered compared to the total number of products:
percentage_ordered = (total_products_ordered / len(products)) * 100

order_status = (total_products_ordered, percentage_ordered)

# 8. Print the order statistics:
print("\nOrder Statistics:")        #order statut : a tuple or list
print(f"Total Products Ordered: {order_status[0]}")
print(f"Percentage of Products Ordered: {order_status[1]:.2f}%")  #order_status[1]: access to the second element
# : starts the format specification.
# .2f means:
#    f → format as a floating-point number,
#   .2 → keep 2 digits after the decimal point.

# 9. Update the inventory by subtracting 1 from the quantity of each ordered product:
for ordered_product in customer_orders:
    if inventory[ordered_product] > 0:
        inventory[ordered_product] -= 1

# 10. Print the updated inventory:
print("\nUpdated Inventory")
for product in products:
    print(f"{product}: {inventory[product]}")
```
